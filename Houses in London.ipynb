Questions and answers for begginer level

```python
# Calculate Price per Square Meter Add a column price_per_sq_meter to calculate the price of the house per square meter.
# Use the formula: Price per Square Meter= Price / Square Meters.
```

```python
df["price_per_sq_meter"] = df.apply(lambda x: x["Price (£)"] / x["Square Meters"], axis = 1)
df
```

```python
# Flag Old and New Buildings Create a column building_age_category: o "New" if Building Age <= 10. o "Old" if 
# Building Age > 10.
```

```python
df["building_age_category"] = df["Building Age"].apply(lambda x: 'New' if x <= 10 else 'Old')
df
```

```python
# Categorise by Garden Availability Add a column garden_flag: o True if Garden is "Yes". o False otherwise.
```

```python
df["garden_flag"] = df["Garden"].apply(lambda x: 'True' if x == 'Yes' else 'False' )
df
```

```python
# Floor Count Category Create a column floor_category: o "Single Floor" if Floors = 1. o "Multi-Floor" if Floors > 1.
```

```python
df["floor_category"] = df["Floors"].apply(lambda x: 'Single Floor' if x == 1 else 'Multi-floor')
df
```

```python
# Intermediate
# Luxury vs Affordable Neighborhoods Create a column neighborhood_category: o "Luxury" if the average Price (£) in 
# the Neighborhood is greater than 2,000,000. o "Affordable" otherwise.
```

```python
df["Neighborhood"].value_counts()
```

```python
avg_price_per_neighborhood = df.groupby("Neighborhood")["Price (£)"].mean()
avg_price_per_neighborhood
```

```python
df["average_price"] = df["Neighborhood"].map(avg_price_per_neighborhood)
df
```

```python
df["neighborhood_category"] = df["average_price"].apply(lambda x: 'Luxury' if x > 2000000 else 'Affordable')
df
```

```python
# Calculate Balcony Accessibility Add a column balcony_accessibility_score:
# o Score = 10 for "High-level Balcony". o Score = 5 for "Low-level Balcony". o Score = 0 for "No Balcony".
```

```python
df["Balcony"].value_counts()
```

```python
df["balcony_accessibility_score"] = df["Balcony"].apply(lambda x: 10 if x == 'High-level Balcony' else 5 if x == 'low-level Balcony' else 0 )
df
```

```python
df['balcony_accessibility_score'] = df['Balcony'].apply(lambda x : 10 if x == 'High-level Balcony' else 5 if x == 'Low-level Balcony' else 0)
df
```

```python
# Material-Based Categorisation Create a column material_category:
#  o "Premium" for houses built with Marble, Granite, or Wood. 
# o "Standard" for other materials.
```

```python
df["Materials"].value_counts()
```

```python
df["material_category"] = df["Materials"].apply(lambda x: 'Premium' if x == 'Marble'or x =='Granite' or x == 'Wood' else 'Standard')
df
```

```python
# Analyse Garage Ownership Add a column garage_flag: o True if Garage is "Yes". o False otherwise.
```

```python
df["Garage"].value_counts()
```

```python
df["garage_flag"] = df["Garage"].apply(lambda x: 'True' if x == 'Yes' else 'False')
df
```
